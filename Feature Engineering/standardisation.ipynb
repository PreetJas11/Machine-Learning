Outliers will remain outliers even after the std. so need to handle them.

Why Std? it effects the result accuracy depending of model used. eg accuracy improved in logistic regression but not in Decision Tree.

When to use? 

- KMeans ( use Euclidean distance measure)
- K Nearest Neighbour ( measure the distance between pairs of samples and these distances are influenced by measurement units)
- Principal Component Analysis (PCA) - to get the feature with max variance
- Artificial NN ( apply gradient Descent)
- Gradient Descent ( theta cal became fast, learning rate in update equ of Gradient descent is same for every param )


> Not required - Decision trees, Random Forest, XGBoost, Gradient Boost

# Effect of Outlier

```python
new_rows = pd.DataFrame({
    'Age': [5, 90, 95],
    'EstimatedSalary': [1000, 250000, 350000],
    'Purchased': [0, 1, 1]
})
```

```python
df = pd.concat([df, new_rows], ignore_index=True)
```

```python
plt.scatter(df['Age'], df['EstimatedSalary'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased', axis=1),
                                                    df['Purchased'],
                                                    test_size=0.3,
                                                    random_state=0)

X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['Age'], X_train['EstimatedSalary'])
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'],color='red')
ax2.set_title("After Scaling")

fig.suptitle('Outliers', fontsize=16)

plt.show()
```
